```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:\\data science class\\assignments\\EDA1_cocl_remain\\EDA1\\Cardiotocographic.csv")
#print(df)

df = data.drop(data.columns[[11, 12, 13]], axis = 1)
print(df)
#1.	Data Cleaning and Preparation:

print(df.isnull().sum())  # finding the emplty cell and how many in one column
column_means = np.round(df.mean(), 5)
mean_dict = column_means.to_dict() #it convert list into dictionary
print(mean_dict)
#fill the blank values with mean of that column
df["LB"].fillna(value = mean_dict['LB'], inplace = True)
df["AC"].fillna(value = mean_dict['AC'], inplace = True)
df["FM"].fillna(value = mean_dict['FM'], inplace = True)
df["UC"].fillna(value = mean_dict['UC'], inplace = True)
df["DL"].fillna(value = mean_dict['DL'], inplace = True)
df["DS"].fillna(value = mean_dict['DS'], inplace = True)
df["DP"].fillna(value = mean_dict['DP'], inplace = True)
df["ASTV"].fillna(value = mean_dict['ASTV'], inplace = True)
df["MSTV"].fillna(value = mean_dict['MSTV'], inplace = True)
df["ALTV"].fillna(value = mean_dict['ALTV'], inplace = True)
df["MLTV"].fillna(value = mean_dict['MLTV'], inplace = True)

df.isnull().sum()
```

```python
# 2.	Statistical Summary

#mean, median, standard deviation,etc all values
print(df.describe())

#IQR = interquartile range = Q3-Q1 
print(df.quantile(0.75)-df.quantile(0.25))

'''DS, DP and NSP columns contain only 0s, indicating no variability.
These columns might need further investigation to understand why they have constant values.'''
```
