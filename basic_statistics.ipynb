```python
#Standardization of Numerical Variables
import numpy as np

mean = df["Avg Price"].mean()
std = df["Avg Price"].std()
ST_avg_Price = (df["Avg Price"]-mean)/std
ST_avg_Price = pd.DataFrame(ST_avg_Price)


mean = df["Total Sales Value"].mean()
std = df["Total Sales Value"].std()
ST_Total_Sales_Value = (df["Total Sales Value"]-mean)/std
ST_Total_Sales_Value = pd.DataFrame(ST_Total_Sales_Value)


mean = df["Discount Rate (%)"].mean()
std = df["Discount Rate (%)"].std()
ST_Discount_Rate = (df["Discount Rate (%)"]-mean)/std
ST_Discount_Rate = pd.DataFrame(ST_Discount_Rate)


mean = df["Discount Amount"].mean()
std = df["Discount Amount"].std()
ST_Discount_Amount = (df["Discount Amount"]-mean)/std
ST_Discount_Amount = pd.DataFrame(ST_Discount_Amount)

mean = df["Net Sales Value"].mean()
std = df["Net Sales Value"].std()
ST_Net_Sales_Value = (df["Net Sales Value"]-mean)/std
ST_Net_Sales_Value = pd.DataFrame(ST_Net_Sales_Value)

df1 = pd.concat([ST_avg_Price,ST_Total_Sales_Value,ST_Discount_Rate,ST_Discount_Amount,ST_Net_Sales_Value], axis =1)
print(df1)
#standardize the column

from sklearn.preprocessing import StandardScaler

df_num = df.iloc[ : , 8:13]

SS = StandardScaler()
SS_X = SS.fit_transform(df_num)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = ["Avg Price", "Total Sales Value", "Discount Rate (%)", "Discount Amount","Net Sales Value"]
print(SS_X)
print("====")
```
