```python
import pandas as pd 
df= pd.read_csv("C:\\data science class\\assignments\\naive and text mining\\blogs.csv")
df
```

```python
df.iloc[1:2,0:1]
```

```python
import nltk
nltk.download("all")
```

```python
# lower
df['Data'] = df['Data'].str.lower()
df.head()
```

```python
#remove all the digits
df['Data'] = df['Data'].str.replace(r'\d+', '', regex=True)
df
```

```python
#Tokenization
from nltk.tokenize import word_tokenize

df['tokenized_text'] = df['Data'].apply(word_tokenize)
df
```

```python
import string
def remove_punctuation(tokens):
    return [i for i in tokens if i not in string.punctuation]

df['punctuation_text'] = df['tokenized_text'].apply(remove_punctuation)
df.head()
```

```python
#Stop words
from nltk.corpus import stopwords
stopwordlist = stopwords.words('english')
stopwordlist

def remove_stopword(tokens):
    return [i for i in tokens if i not in stopwordlist]

df['stopword_text'] = df['punctuation_text'].apply(remove_stopword)
df.head()
```

```python
#Lemmatization or stemming
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_tokens(tokens):
  return [lemmatizer.lemmatize(word) for word in tokens]

df['lemmatized_text'] = df['stopword_text'].apply(lemmatize_tokens)
df
```

```python
# detokenize - after getting required words join again into ine string
def detokenize(tokens):
  return ' '.join(tokens)

df['processed_text'] = df['lemmatized_text'].apply(detokenize)
df.head()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
# Instantiate the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Create a new column for sentiment
df['sentiment'] = df['processed_text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df.head()
df.head(10)
```

```python
df['sentiment0'] = df['sentiment'].apply(lambda x: 'positive' if x > 0 else ('negative' if x<0 else 'neutral'))

print(df["sentiment0"].value_counts())
df.head()
```

```python
Y = df["sentiment0"]
x = df['processed_text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(x)
x_new = X.toarray()

print(x_new.shape)
x_new
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_new, Y)
y_pred = model.predict(x_new)


from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y, y_pred)
import numpy as np
print("accuacy score: ",np.round(accuracy,2))
```
